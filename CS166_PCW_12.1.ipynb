```python
import matplotlib


import pylab as PL
import random as RD
import scipy as SP
import numpy as np

RD.seed()

width = 10
height = 10
initProb = 0.6
pos,neg = 1, -1
T = 2
def init():
    global time, config, nextConfig, T

    time = 0

    config = SP.zeros([height, width])
    for x in range(width):
        for y in range(height):
            if RD.random() < initProb:
                state = pos
            else:
                state = neg
            config[y, x] = state

#     nextConfig = SP.zeros([height, width])

def draw():
    PL.cla()
    PL.pcolor(config, vmin = 0, vmax = 1, cmap = PL.cm.binary)
    PL.axis('image')
    PL.title('t = ' + str(time))
    PL.show()

def step():
    global time, config, nextConfig, T
    
    time += 1
    #update cell
    x, y = RD.randint(0,9), RD.randint(0,9)
    before = config[y,x]
    self = config[y,x]
    neighbors = 0
    mag = 0
    for dx in [-1,1]:
        for dy in [-1,1]:
            neighbors += config[(y+dy)%height, (x+dx)%width]
    
    E = -self*neighbors
    after = -before if RD.random() < min(1,np.exp(-2*E/T)) else before
        
    config[y, x] = after
    print('cell y={},x={} is changed from {} to {}'.format(y,x,before,after))
    
    #calculate magnetization
    mag = 0
    for y in range(height):
        for x in range(width):
            mag += config[y,x]
    mag = mag/(height*width)
    print('magnetization of the system is', mag)

#     config, nextConfig = nextConfig, config

init()
draw()
for _ in range(100):
    print('time', time)
    step()
    print('')
draw()
# import pycxsimulator
# pycxsimulator.GUI().start(func=[init,draw,step])
```
