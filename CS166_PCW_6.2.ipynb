```python
##REVERSED VERSION
import matplotlib

import pylab as plt
import networkx as nx
import random as rd

def initialize():
    global g, active
    g = nx.karate_club_graph()
    g.pos = nx.spring_layout(g)
    for i in g.nodes:
        g.nodes[i]['state'] = 1 if random() < .5 else 0

def observe():
    global g, active
    cla()
    nx.draw(g, vmin = 0, vmax = 1,
            node_color = [g.nodes[i]['state'] for i in g.nodes],
            pos = g.pos)
    plt.show()

def update_randome_edge():
    global g, active
    edge = rd.choice(list(g.edges))
    speaker_index = rd.randint(0,1)
    speaker = edge[speaker_index]
    listener = edge[1-speaker_index]
    g.nodes[listener]['state'] = g.nodes[speaker]['state']
    active = sum([g.nodes[i]['state'] for i in g.nodes])
```

```python
for _ in range(10):
    initialize()
    active_list1 = []
    for _ in range(100):
        update_randome_edge()
    #     observe()
        active_list1.append(active)
    plt.plot(list(range(100)),active_list1)
```
