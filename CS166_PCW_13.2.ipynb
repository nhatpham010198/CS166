```python
def exc2_a(rep):
    import random as rd
    import matplotlib.pyplot as plt
    import numpy as np
    result = [0]
    current = result[-1]
    for i in range(rep-1):
        x = rd.choice([-1,1])
        current += x
        current = -3 if current == -4 else current
        result.append(current)
    plt.cla()
    plt.hist(result,bins = len(np.unique(result)))
    plt.title('histogram')
    plt.show()
    plt.plot(list(range(rep)),result)
    plt.title('simple path')
    plt.show()
    return(result)
exc2_a(1000)
```

```python
def exc2_b(rep):
    import random as rd
    import matplotlib.pyplot as plt
    import numpy as np
    result = [0]
    current = result[-1]
    for i in range(rep-1):
        if current >= 6:
            x = rd.choice([-1,-1,-1,1])
        else:
            x = rd.choice([-1,1])
        current += x
        
        result.append(current)
    plt.cla()
    plt.hist(result,bins = len(np.unique(result)))
    plt.title('histogram')
    plt.show()
    plt.plot(list(range(rep)),result)
    plt.title('simple path')
    plt.show()
    return(result)
exc2_b(1000)
```

```python
def exc4(step,sig):
    import scipy.stats as sts
    import matplotlib.pyplot as plt
    import random as rd
    import numpy as np

    x_axis = [0]
    y_axis = [0]
    current = [x_axis[-1],y_axis[-1]]
    for _ in range(step-1):
        direction = np.random.uniform(0,360)
        dis = rd.gauss(0,sig)
        x = current[0] + dis*math.cos(math.pi/180*direction)
        y = current[1] + dis*math.sin(math.pi/180*direction)
        current = [x,y]
        x_axis.append(x)
        y_axis.append(y)
    plt.cla()
    plt.plot(x_axis,y_axis)
    plt.show
exc4(1000,1)
```

```python
def exc9a(walkers,step):
    import scipy.stats as sts
    import matplotlib.pyplot as plt
    import random as rd
    import numpy as np
    absord = 0
    finish = 0
    for _ in range(walkers):
        x_axis = [0]
        y_axis = [0]
        current = [x_axis[-1],y_axis[-1]]
        holes = [[13,7],[14,7],[13,8],[14,8]]
        for _ in range(step-1):
            dx = rd.choice([-1,0,1])
            dy = rd.choice([-1,0,1])
            x = (current[0]+20+ dx)%41-20
            y = (current[-1]+20+ dy)%41-20
            current = [x,y]
            if current in holes:
                absord += 1
                break
        finish += 1
    finish -= absord
    return(absord,finish)
exc9a(100,1000)
```
