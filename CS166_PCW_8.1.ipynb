Prob of a node not in the LLC is q
Prob of a node in the LLC is 1-q
Total number of nodes is n $(1-q)=n(1-e^{k*(q-1)})$

```python
import numpy as np
def calculate_q(k):
    import numpy as np
    from scipy.optimize import root
    return root(lambda q: q - np.exp(k * (q - 1)), 0).x[0]
k = np.linspace(1,10,100)
y = [calculate_q(i) for i in k]

plt.plot(k,1000*(1-np.array(y)))
plt.show
plt.xlabel('k')
plt.ylabel('Size of LLC')
```
