```python
import matplotlib


import pylab as PL
import random as RD
import scipy as SP
import numpy as np

RD.seed()

width = 20
height = 20
initProb = 0.5
pos,neg = 1, -1
T = 1
def init():
    global time, config, nextConfig, T, mag

    time = 0
    mag = 0

    config = SP.zeros([height, width])
    for x in range(width):
        for y in range(height):
            if RD.random() < initProb:
                state = pos
            else:
                state = neg
            config[y, x] = state
        
            mag += config[y,x]

#     nextConfig = SP.zeros([height, width])

def draw():
    PL.cla()
    PL.pcolor(config, vmin = 0, vmax = 1, cmap = PL.cm.binary)
    PL.axis('image')
    PL.title('t = ' + str(time))
    PL.show()

def step():
    global time, config, nextConfig, T, mag
    
    time += 1
    #update cell
    x, y = RD.randint(0,width-1), RD.randint(0, height - 1)
    before = config[y,x]
    self = config[y,x]
    neighbors = 0
    for (dx,dy) in ([0,1],[0,-1],[-1,0],[1,0]):
        neighbors += config[(y+dy)%height, (x+dx)%width]
    
    E = -self*neighbors
    after = - before if RD.random() < min(1,np.exp(2*E/T)) else before
    mag += after - before
    
    config[y, x] = after





# import pycxsimulator
# pycxsimulator.GUI().start(func=[init,draw,step])
```
